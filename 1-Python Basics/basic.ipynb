```python
## Basic Syntax Rules in Python 
## Case Sensitivity - Python is case-sensitive .

name = "Abhishek"
Name =  "Jadhav"

print(name)
print(Name)
```

```python
age = 45

if age > 40:
    print("You are above 40 years old")
    print(age)  # this line is also indented, so it runs only if age > 40
```

```python
## Multiple Statements in a Single Line
name = "Abhishek"; age = 45; print(name); print(age)
x = 5; y = 10; z = x + y; print(z)
```

```python
variable = 10
print(type(variable))
variable = " Abhishek"
print(type(variable))  # Output: <class 'int'>
```
